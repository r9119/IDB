## Einführung
Das Internet der Dinge (IoT) stellt einen bedeutenden Fortschritt in der Technologie dar und ermöglicht es, dass alltägliche Geräte über das Internet miteinander verbunden und kommunizieren können. In diesem Modul haben wir verschiedene Aspekte des IoT untersucht, wobei der Schwerpunkt auf Techniken zur Datenerfassung lag. Das Hauptziel war es, ein Hardware-System einzurichten, das Umweltdaten sammelt, verarbeitet und basierend auf vordefinierten Bedingungen Feedback gibt. Dieser Bericht beschreibt die durchgeführten Arbeiten zur Erreichung dieses Ziels und hebt die Hardwareeinrichtung, die Methoden zur Datenerfassung, die potenzielle Nutzung eines Raspberry Pi und die während des Projekts aufgetretenen Herausforderungen hervor.

## Hardware Setup
Für dieses Projekt habe ich den bereitgestellten Mikrocontroller und einen CO2-Sensor verwendet, um die Luftqualität in meinem Zimmer zu überwachen. Die Einrichtung umfasste das Anschließen des CO2-Sensors an den Mikrocontroller, der die Luftqualitätsdaten auslas. Basierend auf den Sensormessungen änderte die eingebaute LED am Mikrocontroller ihre Farbe, um das Luftqualitätsniveau anzuzeigen: Grün für gut, Gelb für mäßig und Rot für schlechte Luftqualität. Diese einfache, aber effektive Einrichtung ermöglichte eine Echtzeitüberwachung und Rückmeldung zu den Umgebungsbedingungen in meinem Zimmer.

## Microcontroller Data Collection
Der für die Datenerfassung verwendete Mikrocontroller war ein Adafruit-Modell mit einem WiFi-Airlift-Board. Er wurde so programmiert, dass er in regelmäßigen Abständen Daten vom CO2-Sensor liest. Die Daten wurden dann verarbeitet, um das Luftqualitätsniveau zu bestimmen. Die LED des Mikrocontrollers gab sofortiges visuelles Feedback. Die Daten wurden auch für logging vorbereitet und über das MQTT-Protokoll an eine Thingspeak-Collection gesendet. Diese Einrichtung zeigte die Fähigkeit des Mikrocontrollers, Echtzeit-Datenerfassung und -verarbeitung effizient zu bewältigen.

## Raspberry Pi Differences and uses
Obwohl ich den Raspberry Pi Linux-Computer in meinem Projekt nicht genutzt habe, ist es wichtig zu diskutieren, wie er hätte implementiert werden können. Der Raspberry Pi bietet im Vergleich zum Mikrocontroller mehr Rechenleistung, Speicher und Vielseitigkeit, was ihn zu einem hervorragenden Kandidaten für komplexere IoT-Anwendungen macht.

### Using the Raspberry Pi Like a Microcontroller
Erstens kann der Raspberry Pi auf die gleiche Weise wie der Mikrocontroller zur Datenerfassung und -verarbeitung verwendet werden. Durch das Anschließen des CO2-Sensors an die Pins des Raspberry Pi Grove base hat kann dieser die Luftqualitätsdaten ähnlich wie der Mikrocontroller lesen. Der Raspberry Pi kann auch seine eingebaute LED oder eine externe LED, die angeschlossen ist, verwenden, um basierend auf den Luftqualitätsmessungen visuelles Feedback zu geben, indem die Farben gewechselt werden, um gute, mäßige oder schlechte Luftqualität anzuzeigen. Diese Einrichtung zeigt die Fähigkeit des Raspberry Pi, grundlegende Sensordatenerfassung und sofortiges Feedback ebenso wie der Mikrocontroller durchzuführen.

### Extended Capabilities of the Raspberry Pi
Über das Nachahmen des Mikrocontrollers hinaus kann der Raspberry Pi erweitert werden, um als Datenerfassungsserver für andere Mikrocontroller zu fungieren und als zentraler Knotenpunkt in einem umfangreicheren IoT-Netzwerk zu dienen. Hier ist, wie er genutzt werden kann:

- Datenaggregation und Zentralisierung: Der Raspberry Pi kann Daten von mehreren Sensoren und Mikrocontrollern sammeln. Durch die Einrichtung von MQTT-Verbindungen oder Bluetooth/WLAN mit verschiedenen Mikrocontroller-Einheiten kann er die von ihnen gesendeten Daten aggregieren. Diese zentralisierte Datenerfassung kann das Management und die Überwachung mehrerer Sensoren, die an verschiedenen Standorten verteilt sind, vereinfachen.

- Lokale Datenverarbeitung und Analyse: Mit seiner überlegenen Rechenleistung kann der Raspberry Pi komplexere Datenanalysen durchführen als ein Mikrocontroller. Beispielsweise kann er Trends in den Luftqualitätsdaten im Laufe der Zeit analysieren, Muster identifizieren und sogar maschinelle Lernalgorithmen anwenden, um zukünftige Luftqualitätsniveaus vorherzusagen. Diese Fähigkeit ermöglicht eine anspruchsvollere Verarbeitung und Erkenntnisgewinnung direkt am Rand, wodurch die Notwendigkeit einer ständigen Kommunikation mit der Cloud reduziert wird.

- Speicherlösungen: Der Raspberry Pi kann Daten lokal auf seiner SD-Karte oder einem externen USB-Laufwerk speichern. Diese lokale Speicherung kann als Backup bei Netzwerkausfällen dienen und sicherstellen, dass keine Daten verloren gehen. Sie kann auch historische Daten für langfristige Analysen speichern, was für das Verständnis von Umweltveränderungen über längere Zeiträume hinweg von Vorteil ist.

- Cloud-Integration: Nach der lokalen Verarbeitung und Speicherung der Daten kann der Raspberry Pi auch als Gateway zur Cloud fungieren. Er kann aggregierte und analysierte Daten an Cloud-Dienste zur weiteren Verarbeitung, zum Teilen und zur Visualisierung hochladen. Diese Fähigkeit ermöglicht die Integration von Edge- und Cloud-Computing und nutzt die Vorteile sowohl lokaler als auch cloudbasierter Ressourcen.

- Webserver-Funktionen: Der Raspberry Pi kann einen Webserver hosten, der eine Benutzeroberfläche zur Echtzeitüberwachung der Luftqualitätsdaten bereitstellt. Benutzer können von jedem Gerät im Netzwerk auf diese Schnittstelle zugreifen, was eine bequeme Möglichkeit bietet, Daten zu visualisieren und Benachrichtigungen zu erhalten. Diese Funktion ist besonders nützlich für die Fernüberwachung und das Management von IoT-Geräten.

### Potential Implementation
In einer hypothetischen Implementierung würde der Raspberry Pi als zentraler Knoten in einem Heim- oder Büro-IoT-Netzwerk dienen. Verschiedene Mikrocontroller, die jeweils mit verschiedenen Sensoren (z.B. CO2, Temperatur, Luftfeuchtigkeit) verbunden sind, würden ihre Daten an den Raspberry Pi senden. Der Raspberry Pi würde diese Daten aggregieren, notwendige lokale Analysen durchführen und Echtzeit-Feedback über ein Web-Dashboard bereitstellen. Darüber hinaus würde er regelmäßig Daten mit einem Cloud-Server synchronisieren, um umfassende Langzeitanalysen und Fernzugriff zu ermöglichen. Diese Einrichtung veranschaulicht die Vielseitigkeit und Robustheit, die der Raspberry Pi in eine IoT-Umgebung einbringt, und verbessert sowohl die Fähigkeit als auch die Zuverlässigkeit des Datenerfassungssystems.

Zusammenfassend lässt sich sagen, dass der Raspberry Pi zwar die gleichen grundlegenden Funktionen wie ein Mikrocontroller bei der Datenerfassung ausführen kann, seine erweiterten Fähigkeiten ihn jedoch zu einem leistungsstarken Werkzeug für den Aufbau ausgeklügelter und skalierbarer IoT-Systeme machen. Seine Fähigkeit, Daten zu aggregieren, zu analysieren, zu speichern und weiterzuleiten, macht ihn zu einer integralen Komponente in der modernen IoT-Architektur und überbrückt die Lücke zwischen Edge- und Cloud-Computing.

## Short data analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Import data
data = pd.read_csv('./Data/feeds.csv')

# Convert date to datetime
data['created_at'] = pd.to_datetime(data['created_at'])

# Remove elvation and long/lat columns
data = data.drop(['elevation','longitude','latitude','status'], axis=1)

# Rename field1 to temperature, field2 to humidity, field3 to co2 ppm
data = data.rename(columns={'field1':'temperature','field2':'humidity','field3':'co2'})

# Never really fully ran the microconroller till recently so filter all data before 06-04-2024
data = data[data['created_at'] > pd.to_datetime('2024-06-04T00:00:00+00:00')]

# Combine fields of every 3 rows into one row without 
data = data.groupby(np.arange(len(data))//3).agg({'created_at':'first', 'temperature':'first', 'humidity':'first', 'co2':'first'})

# Handle missing values
data.ffill(inplace=True)

data.head()
```

```python
import plotly.express as px

fig = px.line(data.reset_index(), x='created_at', y='temperature', title='Temperature Over Time')
fig.show()

fig = px.line(data.reset_index(), x='created_at', y='humidity', title='Humidity Over Time')
fig.show()

fig = px.line(data.reset_index(), x='created_at', y='co2', title='CO2 Levels Over Time')
fig.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.seasonal import seasonal_decompose
from sklearn.ensemble import IsolationForest

# Rolling Statistics
data['temp_roll_mean'] = data['temperature'].rolling(window=24).mean()
data['humidity_roll_mean'] = data['humidity'].rolling(window=24).mean()
data['co2_roll_mean'] = data['co2'].rolling(window=24).mean()

# Correlation Heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(data.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()

# Seasonal Decomposition
result = seasonal_decompose(data['temperature'], model='additive', period=24)
result.plot()
plt.show()

# Anomaly Detection with Isolation Forest
clf = IsolationForest(contamination=0.01)
data['anomaly'] = clf.fit_predict(data[['temperature', 'humidity', 'co2']])
data['anomaly'] = data['anomaly'].map({1: 0, -1: 1})

plt.figure(figsize=(14, 6))
plt.plot(data.index, data['co2'], label='CO2')
plt.scatter(data.index, data['co2'], c=data['anomaly'], cmap='coolwarm', label='Anomaly')
plt.legend()
plt.show()
```

```python
from pandas.plotting import lag_plot

# Plot lag plots for the 3 features
fig, axs = plt.subplots(3, 1, figsize=(10, 10))
lag_plot(data['temperature'], ax=axs[0])
axs[0].set_title('Temperature')
lag_plot(data['humidity'], ax=axs[1])
axs[1].set_title('Humidity')
lag_plot(data['co2'], ax=axs[2])
axs[2].set_title('CO2')
plt.tight_layout()
plt.show()
```

```python
sns.histplot(data['temperature'], kde=True)
plt.title('Distribution of Temperature')
plt.show()

sns.histplot(data['humidity'], kde=True)
plt.title('Distribution of Humidity')
plt.show()

sns.histplot(data['co2'], kde=True)
plt.title('Distribution of CO2 Levels')
plt.show()
```
